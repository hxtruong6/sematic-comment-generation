```python
def get_topic_words(topics_list):
    words_list = []
    for topic in topics_list.split('\n'):
        # print(topic)
        tps  = topic.split('"')
        words_list.append([tps[i] for i in range(len(tps)) if i % 2 == 1])
    return words_list
```

```python
words_list_col = []
for idx, row in df.iterrows():
    words_list = get_topic_words(row['topics'])
    # df['sentences'] = 
    # print(words_list)
    words_list_col.append(words_list)
    # for words in words_list:
    #     sen = nlp(words)
    #     print(sen)
    # break
```

```python
df['words_list'] = words_list_col
df.sample(10)
```
